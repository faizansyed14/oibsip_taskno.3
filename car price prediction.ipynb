# Feature engineering


```python
df = df.drop('car_ID', axis =1)
df = df.drop('doornumber', axis =1)
df = df.drop('enginelocation', axis =1)
```

```python
df
```

```python
df.drop("CarName" , axis = 1)
```

```python
df.fueltype.unique()
```

```python
fueltype = pd.get_dummies(df['fueltype']).head()
fueltype
```

```python
fueltype = pd.get_dummies(df['fueltype'], drop_first=True)
fueltype
```

```python
df.aspiration.unique()
aspiration = pd.get_dummies(df['aspiration'])
aspiration.head()
```

```python
aspiration = pd.get_dummies(df['aspiration'], drop_first=True)
aspiration.head()
#0 is turbo and 1 is std
```

```python
carbody = df.carbody
carbody.unique()
carbody = carbody.map({'convertible':0, 'hatchback':1, 'sedan':2, 'wagon':3, 'hardtop':4})
carbody
```

```python
drivewheel = df.drivewheel
drivewheel.unique()
drivewheel = drivewheel.map({'rwd':0, 'fwd':1, '4wd':2})
drivewheel
```

```python
fuelsystem = df.fuelsystem
fuelsystem.unique()
fuelsystem = fuelsystem.map({'mpfi':0, '2bbl':1, 'mfi':2, '1bbl':3, 'spfi':4, '4bbl':5, 'idi':6, 'spdi':6})
fuelsystem
```

```python
new_data = pd.concat([df,fueltype, aspiration, carbody, drivewheel, fuelsystem], axis =1)
```

```python
new_data.head()
```

```python
new_data.info()
```

```python
new_data.drop(['CarName','fueltype','aspiration','carbody','drivewheel','enginetype','cylindernumber','fuelsystem'], axis = 1, inplace= True)
```

```python
new_data.head()
```

```python
new_data.columns
```

```python
new_data.isnull().sum()
```

```python
x = new_data.drop('price', axis = 1)
```

```python
y = new_data[('price')]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x,y ,test_size = 0.3, random_state = 0)
```

```python
print("x_train" , x_train.shape)
print("x_test" , x_test.shape)
print("y_train" , y_train.shape)
print("y_test" , y_test.shape)
```

```python
from sklearn.svm import SVR
model = SVR()
model = model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print("MSE", mean_squared_error(y_test,y_pred))
print("MAE", mean_absolute_error(y_test,y_pred))


```
